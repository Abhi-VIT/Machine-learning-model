```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = {
    'area':[2600, 3000, 3200, 3600, 4000],
    'price' : [550000, 565000, 610000, 680000, 725000]
}
df = pd.DataFrame(data)
df
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
area_df = pd.DataFrame({'area': [3300]})

prediction = reg.predict(area_df)
print(prediction)
```

```python
# the value of M
reg.coef_
```

```python
# the value of intercept b
reg.intercept_
```

```python
#y = m*x + b
135.78767123*3300 + 180616.43835616432
```

```python
%matplotlib inline 
plt.xlabel('area', fontsize= 20)
plt.ylabel('price',fontsize=20)
plt.scatter(df.area,df.price, color = 'red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color = 'blue')
plt.show()
```

```python
score = reg.score(df[['area']], df.price)
score
```

```python
d = pd.read_csv("year.csv")
d.head(3)
```

```python
p = reg.predict(d)
```

```python
p
```

```python
a = df['pcidollar'].to_numpy()
a
```

```python
# Step 1: Calculate SS_residual
SS_residual = np.sum((a - p) ** 2)
SS_residual
```

```python
# Step 2: Calculate SS_total
y_mean = np.mean(a)  # Mean of actual values
SS_total = np.sum((a - y_mean) ** 2)
SS_total
```

```python
#The R² score is calculated as:
#r^2 = 1 - (SSresidual/sstotal)

r_2 = 1-(SS_residual/SS_total)
r_2
```
